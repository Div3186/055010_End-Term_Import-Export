```python
# Composite Measure {Coefficient of Variation | Confidence Interval}

import scipy.stats as stats

sample = np.array([10, 20, 30, 40, 50])

# Mean and Standard Deviation
mean = np.mean(sample)  # Mean of the data
std_dev = np.std(sample)  # Standard deviation of the data

# Coefficient of Variation (CV)
# CV is a measure of relative variability (standard deviation divided by mean)
cv = (std_dev / mean) * 100  # Expressed as a percentage

# Confidence Interval (CI)
# CI provides a range in which the true mean is likely to fall
confidence_level = 0.95  # 95% confidence level
n = len(data)  # Sample size
std_error = std_dev / np.sqrt(n)  # Standard error of the mean

# Z-score for 95% confidence level (from standard normal distribution)
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Confidence interval bounds
ci_lower = mean - z_value * std_error
ci_upper = mean + z_value * std_error
confidence_interval = (ci_lower, ci_upper)

# Output 
print(f"Step 1: The Mean of the data is {mean:.2f} and the Standard Deviation is {std_dev:.2f}.")
print(f"Step 2: The Coefficient of Variation (CV) is {cv:.2f}%, meaning the data varies by {cv:.2f}% relative to the mean.")
print(f"Step 3: The 95% Confidence Interval (CI) for the mean is between {ci_lower:.2f} and {ci_upper:.2f}.")
```

```python
# Inferential for non-Categorical sample

import statsmodels.api as sm

sample1 = np.random.normal(loc=50, scale=10, size=100)
sample2 = np.random.normal(loc=52, scale=12, size=100)

# ===========================
# Test of Mean {t-test | ANOVA}
# ===========================

# 1. t-test (independent samples)
t_stat, p_value_t = stats.ttest_ind(sample1, sample2)
print(f"t-test: t-statistic = {t_stat}, p-value = {p_value_t}")

# 2. ANOVA
f_stat_anova, p_value_anova = stats.f_oneway(sample1, sample2)
print(f"ANOVA: F-statistic = {f_stat_anova}, p-value = {p_value_anova}")

# ============================================
# Test of Variance {F-test | Levene | Bartlett}
# ============================================

# 1. F-test for variances (using variances directly)
f_stat_f = np.var(sample1) / np.var(sample2)
dfn = len(sample1) - 1  # Degrees of freedom for numerator
dfd = len(sample2) - 1  # Degrees of freedom for denominator
p_value_f = 1 - stats.f.cdf(f_stat_f, dfn, dfd)
print(f"F-test: F-statistic = {f_stat_f}, p-value = {p_value_f}")

# 2. Levene's test (for equal variances, robust to non-normality)
w_stat_levene, p_value_levene = stats.levene(sample1, sample2)
print(f"Levene's Test: W-statistic = {w_stat_levene}, p-value = {p_value_levene}")

# 3. Bartlett's test (for equal variances, assumes normality)
b_stat_bartlett, p_value_bartlett = stats.bartlett(sample1, sample2)
print(f"Bartlett's Test: B-statistic = {b_stat_bartlett}, p-value = {p_value_bartlett}")

# =================================================
# Test of Proportion {z-test | Chi-square (Chi-sq)}
# =================================================

# 1. z-test for proportions (example with hypothetical sample)
successes1, successes2 = 30, 40
n1, n2 = 100, 120
z_stat, p_value_z = sm.stats.proportions_ztest([successes1, successes2], [n1, n2])
print(f"Z-test for proportions: Z-statistic = {z_stat}, p-value = {p_value_z}")

# 2. Chi-square test for proportions (example with contingency table)
contingency_table = np.array([[30, 70], [40, 80]])
chi2_stat, p_value_chi2, _, _ = stats.chi2_contingency(contingency_table)
print(f"Chi-square Test: Chi2-statistic = {chi2_stat}, p-value = {p_value_chi2}")

# ==================================
# Test of Correlation {t-test method}
# ==================================

# Pearson correlation coefficient and t-test for correlation significance
corr, p_value_corr = stats.pearsonr(sample1, sample2)
print(f"Pearson Correlation: correlation = {corr}, p-value = {p_value_corr}")

# ============================================
# Test of Normality {Shapiro-Wilk | KS | AD | JB}
# ============================================

# 1. Shapiro-Wilk test
shapiro_stat, p_value_shapiro = stats.shapiro(sample1)
print(f"Shapiro-Wilk Test: W-statistic = {shapiro_stat}, p-value = {p_value_shapiro}")

# 2. Kolmogorov-Smirnov (KS) test
ks_stat, p_value_ks = stats.kstest(sample1, 'norm', args=(np.mean(sample1), np.std(sample1)))
print(f"KS Test: KS-statistic = {ks_stat}, p-value = {p_value_ks}")

# 3. Anderson-Darling test
ad_stat = stats.anderson(sample1)
print(f"Anderson-Darling Test: A2-statistic = {ad_stat.statistic}, critical_values = {ad_stat.critical_values}, significance_level = {ad_stat.significance_level}")

# 4. Jarque-Bera test
jb_stat, p_value_jb = stats.jarque_bera(sample1)
print(f"Jarque-Bera Test: JB-statistic = {jb_stat}, p-value = {p_value_jb}")

# ==============================================
# Non-Parametric Tests {Mann-Whitney | Wilcoxon | Kruskal-Wallis | Friedman}
# ==============================================

# 1. Mann-Whitney U Test (non-parametric alternative to t-test)
u_stat, p_value_u = stats.mannwhitneyu(sample1, sample2)
print(f"Mann-Whitney U Test: U-statistic = {u_stat}, p-value = {p_value_u}")

# 2. Wilcoxon Signed-Rank Test (for paired sample)
# For demonstration, using the same sample1, but typically for paired sample
w_stat, p_value_wilcoxon = stats.wilcoxon(sample1[:50], sample1[50:])
print(f"Wilcoxon Signed-Rank Test: W-statistic = {w_stat}, p-value = {p_value_wilcoxon}")

# 3. Kruskal-Wallis Test (non-parametric alternative to ANOVA)
h_stat, p_value_kruskal = stats.kruskal(sample1, sample2)
print(f"Kruskal-Wallis Test: H-statistic = {h_stat}, p-value = {p_value_kruskal}")

# 4. Friedman Test (for repeated measures, assuming hypothetical 3-sample sample)
sample3 = np.random.normal(loc=48, scale=11, size=100)
friedman_stat, p_value_friedman = stats.friedmanchisquare(sample1[:30], sample2[:30], sample3[:30])
print(f"Friedman Test: chi2-statistic = {friedman_stat}, p-value = {p_value_friedman}")
```
